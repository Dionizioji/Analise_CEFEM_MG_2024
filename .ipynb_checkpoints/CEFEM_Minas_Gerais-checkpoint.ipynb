# Análise da Distribuição CFEM em Minas Gerais - 2024

Este projeto contém uma análise da distribuição da Compensação Financeira pela Exploração de Recursos Minerais (CFEM) em várias cidades de Minas Gerais para o ano de 2024. A análise é realizada por meio de visualizações de dados e estatísticas, utilizando bibliotecas de Python como `pandas`, `matplotlib`, `seaborn` e `plotly`.

## Objetivo

O objetivo deste notebook é fornecer uma visão geral da distribuição da CFEM entre as cidades de Minas Gerais, destacando as principais beneficiárias, a evolução mensal dos valores e algumas estatísticas descritivas.

## Funcionalidades

- **Gráfico de Barras Horizontal**: Exibe as 10 cidades de Minas Gerais com maior valor total distribuído em CFEM no ano de 2024.
- **Gráfico de Linha Temporal**: Mostra a evolução mensal da CFEM ao longo dos meses.
- **Gráfico de Pizza Interativo**: Representa a distribuição da CFEM entre as 5 cidades mais beneficiadas.
- **Análise Estatística**: Valores estatísticos descritivos da CFEM.
- **Mapa de Calor**: Permite uma visualização dos picos dos valores distribuídos, bem como ver o comportamento dos valores arrecadados ao longo dos meses por município.

Os dados foram obtidos em: <a href='https://sistemas.anm.gov.br/arrecadacao/extra/relatorios/distribuicao_cfem_muni.aspx?ano=2024&uf=MG'>Página da Agência Nacional de Mineração</a>



```python
# Importações e configurações
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
from datetime import datetime

# Configurações visuais
plt.style.use('seaborn-v0_8')
colors = px.colors.qualitative.Set3
custom_color_scale = ['#1f77b4', '#2ca02c', '#ff7f0e', '#d62728', '#9467bd',
                     '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']

```

```python
# Funções de formatação
def format_currency_br(value, precision=2):
    """Formata valores monetários no padrão brasileiro"""
    if pd.isna(value) or value == 0:
        return "R$ 0,00"

    if value >= 1_000_000_000:
        return f"R$ {value/1_000_000_000:.{precision}f}B"
    elif value >= 1_000_000:
        return f"R$ {value/1_000_000:.{precision}f}M"
    elif value >= 1_000:
        return f"R$ {value/1_000:.{precision}f}K"
    return f"R$ {value:.2f}"

def clean_currency(x):
    """Limpa e converte valores monetários"""
    if pd.isna(x):
        return 0.0
    if isinstance(x, (int, float)):
        return float(x)
    try:
        x = str(x).strip()
        x = x.replace('\xa0', '')
        x = x.replace(' ', '')
        x = x.replace('.', '')
        x = x.replace(',', '.')
        return float(x) if x else 0.0
    except:
        return 0.0
```

```python
# Carregamento e preparação dos dados
# Carregar o DataFrame
df = pd.read_csv('/home/dionizio/Documentos/ProjetosPraticosPython/CEFEM_Minas_gerais/Dataset.csv')
meses = ['Jan.', 'Fev.', 'Mar.', 'Abr.', 'Mai.', 'Jun.', 'Jul.', 'Ago.', 'Set.', 'Out.']

# Tratamento da coluna Total
df['Total.'] = df['Total.'].apply(clean_currency)

# Tratamento das colunas mensais
for mes in meses:
    df[mes] = df[mes].apply(clean_currency)

```

```python
# Gráfico de Top 10 Cidades
plt.figure(figsize=(12, 6))
top_10 = df.nlargest(10, 'Total.')
plt.barh(top_10['Cidade'], top_10['Total.'])
plt.title('Top 10 Cidades de Minas Gerais - Distribuição CFEM 2024', pad=20)
plt.xlabel('Valor Total (R$)')
plt.ylabel('Cidade')

# Adicionar valores nas barras
for i, v in enumerate(top_10['Total.']):
    plt.text(v, i, format_currency_br(v), va='center')

plt.tight_layout()
plt.show()

```

```python
# Gráfico de Evolução Mensal
plt.figure(figsize=(15, 6))
valores_mensais = df[meses].sum()

plt.plot(range(len(meses)), valores_mensais.values, 'o-', linewidth=2)
plt.fill_between(range(len(meses)), valores_mensais.values, alpha=0.3)
plt.title('Evolução Mensal da Distribuição CFEM 2024')
plt.xticks(range(len(meses)), meses, rotation=45)
plt.ylabel('Valor Total (R$)')

# Adicionar valores nos pontos
for i, v in enumerate(valores_mensais.values):
    plt.text(i, v, format_currency_br(v), ha='center', va='bottom')

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```

```python
# Gráfico de Pizza Interativo
fig = go.Figure()
top_5 = df.nlargest(5, 'Total.')
fig.add_trace(go.Pie(
    labels=top_5['Cidade'],
    values=top_5['Total.'],
    hole=0.4,
    textinfo='label+percent',
    hovertemplate="<b>%{label}</b><br>" +
                  "Valor: %{value:,.2f}<br>" +
                  "Percentual do total distribuído: %{percent}<extra></extra>"
))
fig.update_layout(
    title='Distribuição CFEM - Top 5 Cidades',
    annotations=[dict(text='CFEM 2024', x=0.5, y=0.5, font_size=20, showarrow=False)]
)
fig.show()

```

```python
# Análise Estatística
def generate_statistical_summary(df):
    stats = {
        'Total Distribuído': df['Total.'].sum(),
        'Média por Cidade': df['Total.'].mean(),
        'Mediana': df['Total.'].median(),
        'Desvio Padrão': df['Total.'].std(),
        'Máximo': df['Total.'].max(),
        'Mínimo (não-zero)': df[df['Total.'] > 0]['Total.'].min(),
        'Número de Cidades': len(df[df['Total.'] > 0]),
        'Concentração Top 5': df.nlargest(5, 'Total.')['Total.'].sum() / df['Total.'].sum() * 100,
        'Concentração Top 10': df.nlargest(10, 'Total.')['Total.'].sum() / df['Total.'].sum() * 100
    }

    print("\n=== Análise Estatística CFEM Minas Gerais 2024 ===\n")
    for metric, value in stats.items():
        if 'Concentração' in metric:
            print(f"{metric}: {value:.2f}%")
        elif 'Número' in metric:
            print(f"{metric}: {int(value)}")
        else:
            print(f"{metric}: {format_currency_br(value)}")

generate_statistical_summary(df)

```

```python
# Mapa de Calor Mensal
plt.figure(figsize=(15, 8))
top_10_cidades = df.nlargest(10, 'Total.')
dados_heatmap = top_10_cidades[meses].values

sns.heatmap(dados_heatmap,
            xticklabels=meses,
            yticklabels=top_10_cidades['Cidade'],
            annot=True,
            fmt='.0f',
            cmap='Purples')

plt.title('Distribuição Mensal por Cidade (Top 10) em reais (R$)')
plt.xlabel('Mês')
plt.ylabel('Cidade')
plt.tight_layout()
plt.show()
```
